```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
X = np.random.randn(100, 2)
X = np.hstack((X, np.ones((100, 1))))
y = (X[:, 0] + X[:, 1] + 0.2 * np.random.randn(100) > 0).astype(int)
W = np.array([-1.0, 1.0, 1.0])

plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', edgecolors='k')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Logistic Regression Dataset')

x_vals = np.array([-3, 3])
y_vals = (-W[0] * x_vals - W[2]) / W[1]
plt.plot(x_vals, y_vals, 'k--', label='Original Decision Boundary')

W_train = W.copy()
for epoch in range(1000):
    idx = np.random.randint(0, 100)
    if y[idx] == 0 and np.dot(W_train, X[idx]) >= 0:
        W_train -= 0.01 * X[idx]
    elif y[idx] == 1 and np.dot(W_train, X[idx]) < 0:
        W_train += 0.01 * X[idx]

y_vals_trained = (-W_train[0] * x_vals - W_train[2]) / W_train[1]
plt.plot(x_vals, y_vals_trained, 'c-', label='Trained Decision Boundary')
plt.legend()
plt.show()
```
