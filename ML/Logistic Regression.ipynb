## This is the code for Logistic Regression

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
make_classification
```

```python
dataset = make_classification(n_samples = 100, n_features = 2, n_informative = 2, n_redundant = 0,random_state = 96)
```

```python
X,y = dataset
```

```python
df = pd.DataFrame({"X1":X[:,0], "X2":X[:,1],"Y":y})
```

```python
df
```

```python
plt.scatter(df["X1"],df["X2"],c=y,cmap = 'bwr')
```

```python
X = np.hstack((X,np.ones((100,1))))
```

```python
W = np.array([1.0,1.0,1.0])
y_pred = sigmoid(np.dot(X,W))
epochs = 1000
lr = 0.01
for i in range(epochs):
    y_pred = sigmoid(np.dot(X,W))
    W += (0.01 /len(X))*np.dot(y-y_pred,X)
print(W)
```

```python
plt.scatter(df["X1"],df["X2"],c=y,cmap = 'bwr')
x_vals = np.array([-3,3])
y_vals = (-W[2]-W[0]*x_vals)/W[1]
plt.plot(x_vals,y_vals,'--k')
plt.show()
```
